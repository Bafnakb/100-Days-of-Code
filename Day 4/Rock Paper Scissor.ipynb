```python
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#Write your code below this line ðŸ‘‡
import random

your_choice = int(input("Enter your choice\nRock - 1\nPaper - 2\nScissor - 3\n"))
print("You Chose")
if your_choice == 1:
    print(rock)
elif your_choice == 2:
    print(paper)
elif your_choice == 3:
    print(scissors)


computer_choice = random.randint(1,3)
print("Computer Chose")
if computer_choice == 1:
    print(rock)
elif computer_choice == 2:
    print(paper)
elif computer_choice == 3:
    print(scissors)

if your_choice == 1:
    if computer_choice == 2:
        print("Computer Won")
    elif computer_choice == 3:
        print("You Won")
    else:
        print("Its a Tie")

if your_choice == 2:
    if computer_choice == 1:
        print("You Won")
    elif computer_choice == 3:
        print("Computer Won")
    else:
        print("Its a Tie")

if your_choice == 3:
    if computer_choice == 1:
        print("Computer Won")
    elif computer_choice == 2:
        print("You Won")
    else:
        print("Its a Tie")
```
