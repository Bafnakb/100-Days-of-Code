```python
import random
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

word_list = ["Bafna","Sapna","Manasa","Unni","Aswin","Elephant","Giraffe","India","Mango","Apple"]
chosen_word = random.choice(word_list)
# print(chosen_word)
display = []
for a in chosen_word:
    display.append("_")
print(display)

life = 6
end_game = True
while end_game:
    guess = input("Guess a letter: ").lower()
    count =0   
    for letter in chosen_word:
        if guess == letter.lower():
            display[count] = guess
        count+=1
    if guess not in chosen_word.lower():
        life-=1
    print(stages[life])
        
    print(display)
    
    if "_" not in display:
        print("You won")
        end_game = False
    elif life == 0:
        print(f"You lost, the word was {chosen_word}")
        
        end_game = False

```
